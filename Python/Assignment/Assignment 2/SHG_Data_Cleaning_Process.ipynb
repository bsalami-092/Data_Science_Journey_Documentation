This shows uniqueness. Let me check the uniqueness of values of each categorical data type. I want to know maybe another unique value is due to lower case, space and so on.

```python
Column=['Hotel','Distribution Channel', 'Customer Type','Country', 'Deposit Type','Status','Cancelled']

for i in Column:
    print(f'The unique values in {i} are:')
    print(SHG_Booking[i].unique())
    print('='*50)
```

As I saw the spelling of "Vietnam" as "Viet Nam", I decided to check if there is no Vietnam in this data. Because of the number of assumed unique values Country has, I will only check for Country.

```python
from difflib import SequenceMatcher

categories = SHG_Booking['Country'].unique()
for i, cat1 in enumerate(categories):
    for cat2 in categories[i+1:]:
        ratio = SequenceMatcher(None, str(cat1).replace(' ', '').lower(), str(cat2).replace(' ', '').lower()).ratio()
        if ratio > 0.8 and cat1 != cat2:
            print(f"Potential duplicates: '{cat1}' and '{cat2}' (similarity: {ratio:.2f})")
```

With the result, we can be sure that there is no duplicate. And due to uniqueness confirmed, we can be sure that there are no invalid inputs in the following columns;'Hotel','Distribution Channel', 'Customer Type','Country', 'Deposit Type','Status', and 'Cancelled'.
