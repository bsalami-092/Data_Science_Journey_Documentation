```python
# Changing the data type to the correct data type

House_Rent['Posted On']=pd.to_datetime(House_Rent['Posted On'])

House_Rent
```

```python
# Splitting the Floor column
pattern=r'(.*) out of (.*)'

House_Rent[['House_Floor','Floor_Number']]=House_Rent['Floor'].str.extract(pattern)

House_Rent.head(3)
```

```python
# Deleting the floor column
House_Rent.drop(columns=['Floor'])
```

```python
#-Refresh incase the Floor did not delete
House_Rent.drop(columns=['Floor'],inplace=True)
```
