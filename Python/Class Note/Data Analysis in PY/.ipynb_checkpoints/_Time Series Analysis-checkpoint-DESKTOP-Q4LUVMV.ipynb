```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
House_Rent=pd.read_csv('Cleaned_House_Rent.csv')

House_Rent
```

```python
House_Rent.info()
```

```python
House_Rent=pd.read_csv('Cleaned_House_Rent.csv',index_col=0)

House_Rent
```

```python
# Changing the data type to the correct data type

House_Rent['Posted On']=pd.to_datetime(House_Rent['Posted On'])

House_Rent
```

```python
House_Rent.info()
```

```python
# Time Series Analysis
House_Rent['Year']=House_Rent['Posted On'].dt.year
House_Rent['Month']=House_Rent['Posted On'].dt.month
```

```python
House_Rent.head(2)
```

```python
Monthly_Avg_Rent=House_Rent['Rent'].resample('M').sum()


sns.lineplot(data=House_Rent,y='Rent',x='Year')
plt.figure(figsize=(12,6))
plt.plot(Monthly_Avg_Rent,marker='o')
plt.title('Overall Monthly Rent')
plt.xlabel('Date')
plt.ylabel('Average Rent(All Cities)')
plt.show()
```
