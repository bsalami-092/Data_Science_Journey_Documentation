```python
import torch
import numpy as np
```

```python
Scalar_y=torch.tensor(9)
print(Scalar_y)

Vector_y=torch.tensor([1,12,4])
print(Vector_y)

Matrix_y=torch.tensor([[1,4,3],[4,5,6]])
print(Matrix_y)
```

```python
y=torch.tensor([[1,2,3],[2,6,8],[3,5,7]])

x=torch.tensor([[1,1,4],[5,9,0], [4,5,6]])

print(x.mul(y))
print(x.matmul(y))
```

```python
#Compute the derivative of z=x**3 + 5*y**2 + 4*y + 6
x=torch.tensor(data=4., requires_grad=True)
y=torch.tensor(data=5., requires_grad=True)

z=x**3 + 5*y**2 + 4*y + 6

z.backward()

print(x.grad)
print(y.grad)
```
