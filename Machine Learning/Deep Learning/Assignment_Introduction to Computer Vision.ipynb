```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Load the corrupted image (replace 'corrupted.jpg' with the path to your corrupted image)
corrupted_img = cv2.imread('/Corrupted Student Picture.png')

corrupted_img
```

```python
# Convert from BGR (OpenCV default) to RGB for display
corrupted_img_rgb = cv2.cvtColor(corrupted_img, cv2.COLOR_BGR2RGB)

corrupted_img_rgb
```

```python
# Parameters for contrast and brightness adjustment
alpha = 1.1  # Contrast control (>1 increases contrast)
beta = 11    # Brightness control (+ve brightens)
```

```python
# Correct the image
restored_img = cv2.convertScaleAbs(corrupted_img, alpha=alpha, beta=beta)
restored_img_rgb = cv2.cvtColor(restored_img, cv2.COLOR_BGR2RGB)

restored_img_rgb
```

```python
# Show images side by side for comparison
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.title('Corrupted Image')
plt.imshow(corrupted_img_rgb)
plt.axis('off')

plt.subplot(1,2,2)
plt.title(f'Restored Image\n(alpha={alpha}, beta={beta})')
plt.imshow(restored_img_rgb)
plt.axis('off')

plt.show()
```
