```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.preprocessing import RobustScaler, LabelEncoder
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer

sns.set_style('darkgrid')
sns.set_palette('husl')
```

```python
data=px.data.tips()

data.head()
```

```python
#Encode the dataset
cat_cols=list(data.select_dtypes(include='object').columns)

for i in cat_cols:
    encode=LabelEncoder()
    data[i]=encode.fit_transform(data[i])

data.head()    
```

```python
#Scale the dataset
scaler=RobustScaler()
columns=data.columns

scaled_data=scaler.fit_transform(data)

data=pd.DataFrame(data=scaled_data, columns=columns)

data.head()
```

```python
#Model Training
model=KMeans(n_clusters=3, n_init='auto', random_state=11)
model.fit(data)
preds=model.labels_

preds
```

```python
model.inertia_
```

```python
number_of_clusters=list(range(2,13))
inertia=[]
for clusters in number_of_clusters:
    model=KMeans(n_clusters=clusters, n_init='auto', random_state=11)
    model.fit(data)
    inertia_value=model.inertia_
    inertia.append(inertia_value)


plt.plot(number_of_clusters, inertia, '-o')
plt.title('Elbow Curve for K in the range(2,13)')
```

```python
# firstly install yellowbrick with !pip install yellowbrick

kmeans_model = KMeans()
visualizer = KElbowVisualizer(estimator=kmeans_model, k=(2, 13))
visualizer.fit(data)
visualizer.show()
```
