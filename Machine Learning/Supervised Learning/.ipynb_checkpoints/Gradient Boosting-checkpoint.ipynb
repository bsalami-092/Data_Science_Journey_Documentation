```python
cols = ['Gender', 'Dependents', 'Self_Employed', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History']
test_missing = test_data[train_data[cols].isnull().sum(axis=1) > 2]
print(len(test_missing))
```

```python
for i in ['Gender', 'Dependents', 'Self_Employed']:
    test_data[i].fillna(test_data[i].mode()[0], inplace=True)
```

```python
Y_Rows=train_data[train_data['Loan_Status']=='Y']
N_Rows=train_data[train_data['Loan_Status']=='N']
```

```python
NR_Reduced=Y_Rows.iloc[:198]

Ok_sample=pd.concat([NR_Reduced, N_Rows ],ignore_index=True)

Ok_sample.head()
```

```python
Ok_sample['Loan_Status'].value_counts().plot(kind='bar', color='skyblue')

plt.title('Distribution of Loan_Status')
plt.xlabel('Loan_Status')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
#Encode the categorical variables
Encode_train_cols=list(Ok_sample.select_dtypes(include='object').columns)

for i in Encode_train_cols:
    encoder=LabelEncoder()
    Ok_sample[i]=encoder.fit_transform(Ok_sample[i])


Ok_sample.head()
```

```python
#Encode the categorical variables
Encode_test_cols=list(test_data.select_dtypes(include='object').columns)

for i in Encode_test_cols:
    encoder=LabelEncoder()
    test_data[i]=encoder.fit_transform(test_data[i])


test_data.head()
```

```python
# Split the dataset
X=Ok_sample.drop(columns=['Loan_Status'])
y=Ok_sample['Loan_Status']

# Train and split dataset
X_train, X_test,y_train, y_test=train_test_split(X, y, test_size=0.29, random_state=11, stratify=y)
```

```python
#Scale the dataset

Scaler=RobustScaler()
Columns=X_train.columns

Scaler.fit(X_train)
X_train=Scaler.transform(X_train)
X_test=Scaler.transform(X_test)

#Convert to dataframe
X_train=pd.DataFrame(data=X_train, columns=Columns)
X_test=pd.DataFrame(data=X_test, columns=Columns)

X_train.head()
```

```python
X_test.head()
```

```python
# train the model
model = XGBClassifier(n_estimators = 23,
                      random_state = 11, scale_pos_weight=1)
model.fit(X_train, y_train)
train_preds = model.predict(X_train)
test_preds = model.predict(X_test)
```

```python
cols=['Y', 'N']
conf_train=pd.DataFrame(data=confusion_matrix(y_train,train_preds), columns=cols, index=cols)
conf_test=pd.DataFrame(data=confusion_matrix(y_test,test_preds), columns=cols, index=cols)

sns.set_style('darkgrid')
fig, axes=plt.subplots(nrows=1, ncols=2, figsize=(20,7))
sns.heatmap(data=conf_train, ax=axes[0], cbar=False, annot=True, fmt='d')
sns.heatmap(data=conf_train, ax=axes[1], cbar=False, annot=True, fmt='d')

                      
axes[0].set_xlabel('Actual')
axes[0].set_ylabel('Prediction')
axes[1].set_xlabel('Actual')
axes[1].set_ylabel('Prediction')
```

```python
train_classification=classification_report(y_train, train_preds)
test_classification=classification_report(y_test, test_preds)
print(f'Train Classification: \n{train_classification}')
print(f'Test Classification: \n{test_classification}')
```

# Using Stratified KFold To Tune My Model

```python
from sklearn.model_selection import StratifiedKFold
```

```python
classifier = XGBClassifier(random_state = 11) # init the trainer
st_kfold = StratifiedKFold(n_splits=4, shuffle=True, random_state=11) # init the splitter

split_no = 1
train_scores, test_scores = [],[]
for train_idx, test_idx in st_kfold.split(X=X, y=y):
    # perform the split.
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]
    
    # train the model
    classifier.fit(X_train, y_train)
    train_preds = classifier.predict(X_train)
    test_preds = classifier.predict(X_test)
    
    # evaluate model performance.
    train_f1 = f1_score(y_train, train_preds)
    test_f1 = f1_score(y_test, test_preds)
    train_scores.append(train_f1)
    test_scores.append(test_f1)
    
    print(f'split: {split_no}, train f1: {train_f1}, test f1: {test_f1}')
    split_no += 1
    

print(f'Overall train f1: {sum(train_scores)/len(train_scores)}')
print(f'Overall test f1: {sum(test_scores)/len(test_scores)}')
```

# Hyper-Parameter Optimization

```python
from sklearn.model_selection import GridSearchCV
```

```python
# setup the grid.
param_grid = {
    'n_estimators': [5,10,20,30,40,50],
    'max_depth': [2,3,4,5,6],
    'criterion': ['gini', 'entropy','log_loss']
}

# setup the tuning block

model = GridSearchCV(estimator=classifier, param_grid=param_grid,
                     scoring='f1', return_train_score=True, cv = st_kfold.split(X, y),
                     verbose=3)
model.fit(X=X, y=y)
```
