```python
gender_cols=['basketball', 'football', 'soccer', 'sports', 'softball', 'volleyball', 'swimming', 'cheerleading', 'mall', 
             'shopping', 'clothes', 'dress', 'baseball', 'sports', 'drunk', 'hair', 'drugs']

data['gender']=data.groupby(gender_cols)['gender'].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else 'Unknown'))
```

```python
#Putting bounds on columns containing outliers
data['age']=data['age'].clip(lower=12, upper=22)
data['sex']=data['sex'].clip(lower=0, upper=22)
data['god']=data['god'].clip(lower=0, upper=39)
data['blonde']=data['blonde'].clip(lower=0, upper=16)
```

```python
def cap_outliers(series, lower_quantile=0.01, upper_quantile=0.99):
    lower_bound = series.quantile(lower_quantile)
    upper_bound = series.quantile(upper_quantile)
    return series.clip(lower=lower_bound, upper=upper_bound)

data['NumberOffriends']=cap_outliers(data['NumberOffriends'])
```

```python
#Keeping a another copy of the dataset before encode so as to use in profiling the clusters
cleaned_data=data.copy()
```

```python
#Encode the categorical column
encode=LabelEncoder()
data['gender']=encode.fit_transform(data['gender'])

data.head()    
```

```python
#Scaling the dataset
scaler=StandardScaler()
columns=data.columns #So as to not lose the column title, we store it

scaled_data=scaler.fit_transform(data)

data=pd.DataFrame(data=scaled_data, columns=columns)

data.head()
```

```python
# PCA for 3 components to visualize clusters later
pca = PCA(n_components=3, random_state=11)
columns = ['col_1','col_2','col_3']
data = pca.fit_transform(data)

data = pd.DataFrame(data=data, columns=columns)
data.head()
```

```python
#Determining the optimal number of clusters for the Kmeans Model

inertia = []
sil_scores = []
number_of_clusters = range(2, 11)

for k in number_of_clusters:
    K_model=KMeans(n_clusters=k, n_init='auto', random_state=11)
    preds = K_model.fit_predict(data)
    inertia.append(K_model.inertia_)
    sil = silhouette_score(data, preds)
    sil_scores.append(sil)

plt.figure(figsize=(12,5))
plt.subplot(1, 2, 1)
plt.plot(number_of_clusters, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal k')


plt.subplot(1, 2, 2)
plt.plot(number_of_clusters, sil_scores, 'bx-')
plt.xlabel('k')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores For k')

plt.show()

print(f'Best k based on silhouette score: {number_of_clusters[np.argmax(sil_scores)]}')
```

```python
visualizer = KElbowVisualizer(estimator=K_model, k=(2, 11))
visualizer.fit(data)
visualizer.show()
```

```python
#Setup KMeans
K_model = KMeans(n_clusters=5, n_init="auto", random_state=11)
K_model.fit(data)
K_labels= K_model.labels_
```

```python
#Setup DBSCAN
DB_model=DBSCAN(eps=1.0, min_samples=6)
DB_model.fit(data)
DB_labels=DB_model.labels_

DB_labels
```

```python
#Setup AgglomerativeClustering
Ag_model=AgglomerativeClustering(n_clusters=10, linkage='ward')
Ag_model.fit(data)
Ag_labels = Ag_model.labels_

Ag_labels
```

```python
print(f'The inertia for the KMeans Model is:\n', K_model.inertia_)
print('='*50)
print(f'The silhoutte score for the KMeans Model is:\n', silhouette_score( data, K_labels) )  
print('='*50)
print(f'The silhoutte score for the DBSCAN Model is:\n',silhouette_score(X=data, labels=DB_labels))
print('='*50)
print(f'The silhoutte score for the AgglomerativeClustering Model is:\n',silhouette_score(X=data, labels=Ag_labels))
```

```python
px.scatter_3d(data_frame=data, x='col_1', y='col_2', z='col_3',
              color = K_labels)
```

```python
cleaned_data['clusters'] = [f'cluster_{x}' for x in K_labels]
cleaned_data['gender_encoded']=encode.fit_transform(cleaned_data['gender'])

cleaned_data.head()
```

```python
tsne = TSNE(n_components=2, random_state=11)
X_tsne = tsne.fit_transform(data)

plt.figure(figsize=(10,6))
sns.scatterplot(x=X_tsne[:, 0], y=X_tsne[:, 1], hue=cleaned_data['clusters'], palette='tab10', s=30)
plt.title('t-SNE Cluster Visualization')
plt.xlabel('t-SNE Dimension 1')
plt.ylabel('t-SNE Dimension 2')
plt.legend(title='Cluster')
plt.show()
```

```python
cluster_interest_means=cleaned_data.groupby('clusters')[Interest_cols].mean()

# Display the top interests per cluster (top 5 highest mean features)
print("Top 5 Interest Terms Defining Each Cluster:")
for cluster_id in cluster_interest_means.index:
    top_interests=cluster_interest_means.loc[cluster_id].sort_values(ascending=False).head(5)
    print(f"\nCluster {cluster_id}:\n", top_interests)

# Display demographic summary per cluster
cluster_demo_means=cleaned_data.groupby('clusters')[['age', 'gender_encoded', 'gradyear', 'NumberOffriends']].mean()

print("\nDemographic Mean Values for Each Cluster:\n", cluster_demo_means)
cluster_demo_means

# Display gender distribution per cluster
gender_counts=cleaned_data.groupby(['clusters', 'gender']).size().unstack(fill_value=0)

print("\nGender Counts per Cluster:\n", gender_counts)
```
