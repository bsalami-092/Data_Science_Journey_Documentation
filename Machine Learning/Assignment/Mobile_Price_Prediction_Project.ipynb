```python
#Scale the dataset
X=train_data.drop(columns=['price_range'])
y=train_data['price_range']


#Split the scaled data 
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.30, random_state=11, stratify=y) 
#No need to use stratify since my target data looks very balanced but to be on a safer side
```

```python
scaler=StandardScaler()
Columns=X_train.columns #so as to keep our column names

X_train=scaler.fit_transform(X_train)

#Convert to Dataframe
X_train=pd.DataFrame(data=X_train, columns=Columns)
X_train.head()
```

```python
#Model Training
models={
    'Random Forest': RandomForestClassifier(random_state=11, n_estimators=100, max_depth=5),
    'Logistic Regression': LogisticRegression(multi_class='multinomial', max_iter=1000, random_state=11),
    'SVM': SVC(kernel='rbf', probability=True, random_state=11),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=11)
}

st_kfold=StratifiedKFold(n_splits=4, shuffle=True, random_state=11)

for name, model in models.items():
    split_no=1
    train_scores, test_scores=[], []
    
    for train_idx, test_idx in st_kfold.split(X, y):
        #This performs the split. Not caring if the data X and y is scaled or not. It only cares about the proportion not the values in X and y
        #But because we have scaled the data before, no need again
        X_train, X_test=X.iloc[train_idx], X.iloc[test_idx]
        y_train, y_test=y.iloc[train_idx], y.iloc[test_idx]
        
        # train the model
        model.fit(X_train, y_train)
        
        # predictions
        train_preds=model.predict(X_train)
        test_preds=model.predict(X_test)
        
        # evaluate
        train_f1=f1_score(y_train, train_preds, average='weighted')
        test_f1=f1_score(y_test, test_preds, average='weighted')
        train_scores.append(train_f1)
        test_scores.append(test_f1)
        
        print(f'Model: {name}, split: {split_no}, train f1: {train_f1:.4f}, test f1: {test_f1:.4f}')
        split_no += 1
    
    print(f'Overall {name} train f1: {sum(train_scores)/len(train_scores):.4f}')
    print(f'Overall {name} test f1: {sum(test_scores)/len(test_scores):.4f}')
    print("="*60)
```

```python
#Using GridSearchCV
RF_model=RandomForestClassifier(random_state=11)

RF_param_grid={
    'n_estimators': [10, 50, 100],
    'max_depth': [2, 4, 6, None],
    'criterion': ['gini', 'entropy', 'log_loss']
}

RF_grid=GridSearchCV(
    estimator=RF_model,
    param_grid=RF_param_grid,
    scoring='f1_weighted',
    cv=st_kfold,
    verbose=2,
    n_jobs=-1
)

RF_grid.fit(X, y)
print("Random Forest Best Params:", RF_grid.best_params_)
print("Random Forest Best F1:", RF_grid.best_score_)
```

```python
LR_model=LogisticRegression(multi_class='multinomial', max_iter=2000, random_state=11)

LR_param_grid={
    'C': [0.01, 0.1, 1, 10, 100],  # Regularization strength
    'solver': ['lbfgs', 'saga']
}

LR_grid=GridSearchCV(
    estimator=LR_model,
    param_grid=LR_param_grid,
    scoring='f1_weighted',
    cv=st_kfold,
    verbose=2,
    n_jobs=-1
)

LR_grid.fit(X, y)
print("Logistic Regression Best Params:", LR_grid.best_params_)
print("Logistic Regression Best F1:", LR_grid.best_score_)

```

```python
SV_model=SVC(probability=True, random_state=11)

SV_param_grid={
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf', 'poly'],
    'gamma': ['scale', 'auto']
}

SV_grid=GridSearchCV(
    estimator=SV_model,
    param_grid=SV_param_grid,
    scoring='f1_weighted',
    cv=st_kfold,
    verbose=2,
    n_jobs=-1
)

SV_grid.fit(X, y)
print("SVM Best Params:", SV_grid.best_params_)
print("SVM Best F1:", SV_grid.best_score_)
```

```python
XG_model=XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=11)

XG_param_grid={
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.3],
    'subsample': [0.8, 1.0]
}

XG_grid=GridSearchCV(
    estimator=XG_model,
    param_grid=XG_param_grid,
    scoring='f1_weighted',
    cv=st_kfold,
    verbose=2,
    n_jobs=-1
)

XG_grid.fit(X, y)
print("XGBoost Best Params:", XG_grid.best_params_)
print("XGBoost Best F1:", XG_grid.best_score_)
```

```python
#Model Evaluation function with plots
def model_evaluation(model, X_train, X_test, y_train, y_test, model_name='Model'):
    y_pred=model.predict(X_test)
    f1=f1_score(y_test, y_pred, average='weighted')
    print(f'\n{model_name} - Weighted F1-score: {f1:.4f}\n')
    print(f'\nClassification Report:\n{classification_report(y_test, y_pred)}\n')
    # Confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(6,3))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
                xticklabels=np.unique(y_test),
                yticklabels=np.unique(y_test))
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title(f"{model_name} - Confusion Matrix")
    plt.show()
```

```python
model_evaluation(RF_grid.best_estimator_, X_train, X_test, y_train, y_test, "Random Forest")
```

```python
model_evaluation(LR_grid.best_estimator_, X_train, X_test, y_train, y_test, "Logistic Regression")
```

```python
model_evaluation(SV_grid.best_estimator_, X_train, X_test, y_train, y_test, "SVM")
```

```python
model_evaluation(XG_grid.best_estimator_, X_train, X_test, y_train, y_test, "XGBoost")
```

```python
Results = {
    "Random Forest": RF_grid.best_score_,
    "Logistic Regression": LR_grid.best_score_,
    "SVM": SV_grid.best_score_,
    "XGBoost": XG_grid.best_score_
}

pd.DataFrame(Results.items(), columns=["Model", "CV Weighted F1"])
```
